```python
import numpy as np 
import pandas as pd
```

```python
data = pd.read_csv("../input/bearingfeatures/features.csv")
```

```python
data.describe()
```

```python
x = data.drop(data.columns[[0, 4]], axis = 1)
y = data['labels']
x.shape, y.shape
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.1, random_state=4)
```

```python
y_train.shape, x_train.shape
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score 
```

```python
classifier = SVC(kernel = 'linear', random_state=0)
classifier.fit(x_train, y_train)
```

```python
y_pred= classifier.predict(x_test)
```

```python
m = {9: 'Inner Raceway Fault', 11: 'Outer Raceway Fault Orthogonal to Load', 12:'Outer Raceway Fault in Load Zone', 13: 'No Fault'}
result = pd.Series(y_pred).map(m)
truth_value = (abs(y_pred - y_test) == 0)
result
```

```python
accuracy_final = accuracy_score(y_test, y_pred)
accuracy_final
```

```python
from sklearn.metrics import confusion_matrix 
cm = confusion_matrix(y_test, y_pred)
```

```python
target_names = m.values()
import matplotlib.pylab as plt
import itertools
plt.imshow(cm, cmap=plt.get_cmap('Blues'))
plt.title("Confusion Matrix for Model")
if target_names is not None:
    tick_marks = np.arange(len(target_names))
    plt.xticks(tick_marks, target_names, rotation = 90)
    plt.yticks(tick_marks, target_names)
thresh = cm.max() / 2
for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
    plt.text(j, i, "{:,}".format(cm[i, j]), horizontalalignment="center",color="white" if cm[i, j] > thresh else "black")
plt.figure(figsize=(8, 6))
plt.show()
```
