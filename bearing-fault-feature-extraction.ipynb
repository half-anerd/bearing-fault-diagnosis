```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("../input/48k-drive-end-faults/data.csv")
data = data.drop(data.columns[[0]], axis = 1)
data
```

```python
x = data.drop(['labels'], axis = 1)
```

```python
peak_f = []
peak_A = []
sum_f = []
```

```python
Fs = 48000
n = 3600
k = np.arange(n)
Ts = n/Fs
frq = k/Ts 
frq = frq[range(int(n))]
frq[:5]
```

```python
for i in range(len(x)):
    X = np.array(x[i:i+1])
    X = np.fft.fft(X)
    peak_A.append(np.max(np.abs(X)))
    peak_f.append(frq[np.argmax(np.abs(X))])
    S = np.abs(X**2)/n
    sum_f.append(np.sum(S))

print(len(peak_f), len(peak_A), len(sum_f))
```

```python
peakFrequency = np.array(peak_f)
peakAmplitude = np.array(peak_A)
powerSum = np.array(sum_f)
features = pd.DataFrame({'peakFrequency': peakFrequency, 'peakAmplitude': peakAmplitude, 'powerSum': powerSum, 'labels' : data['labels']}, 
                        columns=['peakFrequency', 'peakAmplitude', 'powerSum', 'labels'])
features.head()
```

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize = (10, 5))

# creating the bar plot
plt.bar(data['labels'].squeeze(), features["powerSum"], color ='maroon',
        width = 0.4)
plt.show()
plt.bar(data['labels'].squeeze(), features["peakFrequency"], color ='maroon',
        width = 0.4)
plt.show()
plt.bar(data['labels'].squeeze(), features["peakAmplitude"], color ='maroon',
        width = 0.4)
plt.show()
```

```python
features.to_csv("features.csv")
```
